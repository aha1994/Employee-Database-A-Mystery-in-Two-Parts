```python
# Dependencies
import pandas as pd
import psycopg2
import matplotlib.pyplot as plt
import sqlalchemy
```

```python
# make a connection to the database
from sqlalchemy import create_engine
engine = create_engine('postgresql://fakeuser:fakepass@localhost:5432/Employee Database: A Mystery in Two Parts - AHA')
connection = engine.connect()

# read a sql query into a dataframe
salaries = pd.read_sql('select salaries.emp_no, salaries.salary,titles.title from salaries inner join titles on salaries.emp_no = titles.emp_no', connection)
```

```python
salaries.head()
```

```python
# dataframe manipulation 
titleGroup = salaries.groupby('title')
titleGroup['salary'].mean()
```

```python
# matplotlib plot generation, bar chart of avg salary by potition
x_axis = salaries['title'].unique()
y_axis = titleGroup['salary'].mean()

plt.bar(x_axis, y_axis, color='orange')
plt.title('Average Salary by Job Title')
plt.xlabel('Job Position')
plt.ylabel('Average Salary($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('Average Salary by Job Title.png')
plt.show()
```
